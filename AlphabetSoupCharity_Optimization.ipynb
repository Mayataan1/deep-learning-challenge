```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
y = dummies["IS_SUCCESSFUL"].values
X = dummies.drop(["IS_SUCCESSFUL"],1).values

# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
y2 = dummies2["IS_SUCCESSFUL"].values
X2 = dummies2.drop(["IS_SUCCESSFUL"],1).values

# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create a StandardScaler instances
scaler2 = StandardScaler()

# Fit the StandardScaler
X2_scaler = scaler2.fit(X2_train)

# Scale the data
X2_train_scaled = X2_scaler.transform(X2_train)
X2_test_scaled = X2_scaler.transform(X2_test)
```

## Compile, Train and Evaluate the Model

Initial Model two layers relu/relu 8/5

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE  (add class activity reference here, pulled from ClassStarterCode.ipynb saved to this resources directory)
number_input_features = len(X_train[0])
hidden_nodes_layer1 =  8
hidden_nodes_layer2 = 5

nn = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model anc create a callback that saves the model's weights every five epochs
# https://keras.io/api/callbacks/
# https://keras.io/api/callbacks/model_checkpoint/
checkpoint_filepath = '/Resources/checkpoint-1'

my_callbacks = tf.keras.callbacks.ModelCheckpoint(    
    filepath=checkpoint_filepath,
    save_weights_only=True,
    monitor='val_accuracy',
    mode='auto',
    save_freq = 5)


fit_model = nn.fit(X_train_scaled,y_train,epochs=100, callbacks=my_callbacks)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")

# # Export our model to HDF5 file
#  YOUR CODE GOES HERE
nn.save('DeepLearningModelnn.h5')
```
